# Importing the Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

# Loading the Dataset

```python
df = pd.read_csv("force2020_data_unsupervised_learning.csv", index_col = "DEPTH_MD")
```

```python
df.dropna(inplace=True)
```

# Transform the Data

```python
df.describe()
```

```python
scaler = StandardScaler()
```

We the fit the standard scaler to the data and then transform our data using that scaler

```python
df[['RHOB_T','NPHI_T','GR_T','PEF_T','DTC_T']]=scaler.fit_transform(df[['RHOB','NPHI','GR','PEF','DTC']])
```

```python
df
```

# Identifying Optimum Number of Clusters

```python
# Create function to work out optimum number of clusters
def optimise_k_means(data,max_k):
    means=[]
    inertias=[]
    for k in range(1,max_k):
        kmeans=KMeans(n_clusters=k)
        kmeans.fit(data)
        means.append(k)
        inertias.append(kmeans.inertia_)
    #Generate the elbow plot
    fig=plt.subplots(figsize=(10,5))
    plt.plot(means,inertias,'o-')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()
```

```python
optimise_k_means(df[['RHOB_T','NPHI_T']],10)
```

# Applying K Means Clustering

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(df[['NPHI_T','RHOB_T']])
```

```python
df['kmeans_1']=kmeans.labels_
```

```python
for i in range(1, 6):
    kmeans = KMeans(n_clusters=i)
    df[f'KMeans_{i}'] = kmeans.fit_predict(df[['NPHI', 'RHOB']])
```

```python
df
```

# Plotting The Results

```python
plt.scatter(x=df['NPHI'],y=df['RHOB'],c=df['KMeans_5'])
plt.xlim(-0.1,1)
plt.ylim(3,1.5)
plt.show()
```

# Creating Multiple Clusters

```python
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(20, 5))

for i, ax in enumerate(fig.axes, start=1):
    ax.scatter(x=df['NPHI'],y=df['RHOB'],c=df[f'KMeans_{i}'])
    ax.set_ylim(3,1.5)
    ax.set_xlim(0,1)
    ax.set_title(f'N Clusters: {i}')
```
